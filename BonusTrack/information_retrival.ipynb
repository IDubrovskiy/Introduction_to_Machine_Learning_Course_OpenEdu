```python
import pandas as pd
import math
```

```python
sample_text = ['holder ha it attack when the holder attempt to use a move all it other move are disabled until it leaf battle or loses this item the restriction end even if this item is swapped for another choice item via trick or switcheroo'
,'experience is split across two group pokemon who participated in battle and pokemon holding this item each pokemon earns experience a though it had battled alone divided by the number of pokemon in it group then divided by the number of group pokemon holding this item who also participated in battle effectively earn experience twice fainted pokemon never earn experience and empty group are ignored thus if a single pokemon is holding this item and the only pokemon who battled faint from explosion the holder will gain full experience'
,'damaged the holder for it max hp when the holder is struck by a contact move damage the attacker for it max hp if the attacker is not holding an item it will take this item'
,'trainer will skip encounter with wild pokemon of a lower level than the lead party pokemon this effect wear off after the trainer take step'
,'at the end of each turn if any of the holder stats have a negative stat modifier the holder consumes this item to remove the modifier from those stats']
```

```python
from nltk.tokenize import word_tokenize
sentences = []
word_set = []

for sent in sample_text:
    words = [word.lower() for word in sent.split(sep=' ') if word.isalpha()]
    sentences.append(words)
    for word in words:
        if word not in word_set:
            word_set.append(word)
# Set of words
word_set = set(word_set)
# total documents in our corpus
total_docs = len(sample_text)
print('Total documents: ', total_docs)
print('Total words: ', len(word_set))
```

```python
tf_dict = {}
sent = sentences[3]
sent_len = len(sent)
for word in sent:
    tf_dict[word] = 0
for word in sent:
    tf_dict[word] += 1
for word in tf_dict.keys():
    tf_dict[word] = tf_dict[word] / sent_len
```

```python
idf_dict = {}
for word in word_set:
    idf_dict[word] = 0
for word in word_set:
    for sentence in sentences:
        if word in sentence:
            idf_dict[word] += 1
for word in idf_dict.keys():
    idf_dict[word] = math.log(5 / idf_dict[word], 2)
```

```python
tf_idf_dict = {}
for word in sent:
    tf_idf_dict[word] = tf_dict[word] * idf_dict[word]
```

```python
df = pd.DataFrame(tf_idf_dict.values(), tf_idf_dict.keys(), columns=["TF-IDF"])
df = df.sort_values('TF-IDF', ascending=False)
```
